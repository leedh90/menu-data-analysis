```python
for i, item in enumerate(item_count.iteritems(), 0):
    print("TOP", i + 1, ":", item)
# enumerate는 인덱스와 값을 동시에 꺼낼 수 있다.
# 첫번째는 값을 item에 넣어주고, 1는 인덱스가 i 인덱스가 0부터 시작한다는 의미
# iteritems는 튜플형태로 값을 꺼내온다.
# item에는 시리즈의 index와 values를 tuple로 반복해서 써내온다.
```

```python
# 메뉴별 주문개수와 총량!
for i, item in enumerate(item_count.iteritems(), 0): # 0은 인덱스의 시작값
    print("TOP", i + 1, ":", item[0], item[1])
    # item[0]은 인덱스, item[1]은 values
```

```python
# 많이 팔린 top10까지느 메뉴별 주문 개수!
for i, (pre, post) in enumerate(item_count.iteritems(), 0): # 0은 인덱스의 시작값
    # iteritems는 튜플형식이기 때문에 (변수1, 변수2) 이런 튜플형태로도 받을 수 있다.
    print("TOP", i + 1, ":", pre, post)
```

```python
# 시리즈의 모든 row(행)에 있는 str을 추출해서 $를 제거(인덱스 슬라이싱)하고 float타입으로 변경해준다
# 시리즈를 모두 변경하기 위해서는 for문 반복문을 돌려야함
# float()은 비파괴 함수로 원본을 변경시키지 않는다.
# float()한 결과를 리스트에 모아서 반복문이 끝나면 리스트를 다시 df의 컬럼에 집어 넣는다.(덮어쓰기)
result = []
for x in food["item_price"]: # x = item_price의 각 인덱스에있는 값을 말한다.
    result.append(float(x[1:]))
result[:10]
```

```python
salad_bowl = 0 # 몇번 주문됫는지 구하는 변수
for x in range(len(food)): #전체 주문 내역을 다 돌리고 그중에서 동일한 것만 카운트
    if food["item_name"][x] == "Veggie Salad Bowl":
        salad_bowl = salad_bowl + food["quantity"][x]
salad_bowl # Veggie Salad Bowl의 총 주문 횟수를 구할 수 있다.
```

```python
chicken_conunt1 = 0
for x in food[food['item_name'] == 'Chicken Bowl'].index:
    # 치킨보울이 있는 인덱번호의 퀀티티 값을 비교한다.
    if food["quantity"][x] >= 2: # 수량이 2 이상인것만 찾아
        chicken_conunt1 = chicken_conunt1 + 1 # 주문횟수를 1씩 올려줘라
chicken_conunt1 # 출력
```
